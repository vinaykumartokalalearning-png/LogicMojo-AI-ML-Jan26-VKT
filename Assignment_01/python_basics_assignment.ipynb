```python
# Leap year check

year = int(input("enter your year"))

if(year%4 == 0 and year%100 != 0) or (year%400 == 0):
    print("Leap year")

else:
    print("not a leap year")
```

```python
# Grade from Marks

var = int(input("enter your marks"))

if (var>=90 and var<=100):
    print("A")

elif (var>=80 and var<90):
    print("B")

elif (var>=70 and var<80):
    print("C")

elif (var>=60 and var<70):
    print("D")

elif (var>=0 and var<60):
    print("F")

else:
    print("enter valid marks")
```

```python
# Count Digits

n = int(input("enter a number"))
digit_count = len(str(abs(n)))
print(digit_count)
```

```python
# Sum of Digits

n = int(input("enter a number"))
s = str(n)
total = 0
for i in s:
    total += int(i)
print("sum of digits is:", total)
```

```python
# Reverse a String

text = "python"
reverse = ""
for i in range(len(text)-1, -1, -1):
    reverse += text[i]

print("text:", text)
print("reversed:", reverse)
```

```python
# Palindrome String

text = str(input("eneter a text"))
reverse = ""
for i in range(len(text)-1, -1, -1):
    reverse += text[i]
    
if (text == reverse):
    print("Palindrome")
else:
    print("Not a Palindrome")
```

```python
# Count Vowels

text = str(input("eneter a text"))
vowels = "aeiouAEIOU"
count = 0
for ch in text:
    if ch in vowels:
        count += 1
print("vowels:", count)
```

```python
# Count Words

sentence = str(input("eneter a sentence"))
words = sentence.split()
word_count = len(words)
print("total number of words:", word_count)
```

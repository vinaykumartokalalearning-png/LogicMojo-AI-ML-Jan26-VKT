# Python Basics for ML

## Lists

```python
# example 21 - create + index

nums = [10,20,30,40,50]
print(f"example 21: {nums[0]} {nums[-1]}")
```

```python
# example 22 -slicing
print(f"example 22: {nums[1:3]} {nums[:2]} {nums[::2]}")
```

```python
# example 23 append vs extend

xs = [1,2]
xs.extend([4,5]) # list length increases by the number of elements in the iterable
xs.append([3,6]) # list length always increases by exactly one
print(xs)
```

```python
#example 24 insert

xs.insert(1,99)
print(xs)
```

```python
# example 25 pop returns the removed element

last = xs.pop()
print(xs)
```

```python
# example 26 remove deletes first matching value
xs.remove(99)
print(xs)
```

```python
# example 27 - sorted() vs list.sort()

ys = [4,2,3,1]

print("example 27:", sorted(ys), ys)

ys.sort()

print("example 27:", ys)
```

```python
# Example 28: list comprehension (map + filter in one)

squares_of_even = [n*n for n in range(10) if n%2==0]
print("Example 28:", squares_of_even)
```

```python
# Example 29: aliasing vs copying
a= [1,2,3]
b = a # aliasing
c = a.copy() # shallow copy
b.append(4)
print(a,b,c)
```

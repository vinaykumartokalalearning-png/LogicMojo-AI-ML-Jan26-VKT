### Python Basics

```python
# Variables - names for values

name = "Vinay"
age = 31
is_employee = True
type(name), type(age), type (is_employee)  # Sring, integer, Boolean
```

```python
#### Core data types with examples

#1 Numbers: int, float

x = 10 # int
y = 3.14 # float
total = x + y
print(total)

#2 Boolean: bool

is_employee = age >= 24
print(is_employee) # True or False

# #3 Strings: Str

text = " Hello AI "
print(text.strip().lower())  # "hello ai"

city = "Hyderabad"
print(f"my city is {city}") # variable city passed in curly braces, curly braces defines blocks of code
```

```python
### Collections (most used in data work)

# 1 List (ordered, mutable) - use list when you need an ordered collection you may change

scores = [10,20,30] # Square brackets with ordered collection list
scores.append(40) # paentheses for function parameters or group of math terms
print(scores)

#2 indexing and slicing

nums = [5,6,7,8,9]
print(nums[0]) # first item
print(nums[-1]) # last item
print(nums[1:4]) # items at index 1,2,3

#3 Tuple (ordered, immutable) - use tuples when values should not change

point = (10,20)
print(point[0],point[1])

# 4 Dict (key value mapping) - use dictionaries to store labeled data

student = {"name" : "Vinay", "age" : 30, "City" : "Bangalore"}
print(student["name"])
student["age"] = 31
print(student["age"])

# set (unique values) - use sets to remove duplicate values

ids = [101,101,102,103,103]
unique_ids = set(ids)
print(unique_ids)
```

```python
#### Control Flow (if/for/while)

#1 If condition

score = 70
if score >= 60:
    print("pass")

else:
    print("fail")

#2 for loop

names = ["Aravind", "Bala", "Charan"]
for n in names:
    print(n)

for i,n in enumerate(names, start=1):  #enumerate gives index + value:
    print(i,n)

#3 Functions - reusable logic

def greet(name: str) -> str:
    return f"Hello, {name}!"

print(greet("Aravind"))

# Why functions matter:

    # Less copy-paste
    # Easier testing
    # Cleaner notebooks/scripts

```
